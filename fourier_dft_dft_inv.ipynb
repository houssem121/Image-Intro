```python
import cmath
import numpy
import numpy as np
import math
import matplotlib.pyplot as plt
import matplotlib as mpl

def dft(f):
    N=int(len(f))
    mm=[]
    for k in range(N):
        F=0
        for n in range(N):
            v=2*np.pi*n*k*1j/N
            F=F+f[n]*np.exp(-v)
        mm.append(F/N)

    return mm



def dft_1(f):
    N=int(len(f))
    mm=[]
    for k in range(N):
        F=0
        for n in range(N):
            v=2*np.pi*n*k*1j/N
            F=F+f[n]*np.exp(v)
        mm.append(F)


x=np.linspace(0,10000,4000)
y=100*(np.cos(2*np.pi*8000*x)*((5*np.cos(400*np.pi*x)+5*np.cos(200*np.pi*x))+1))
x1=np.linspace(0,10000,4000)
plt.figure(figsize = (10, 6))
plt.stem(x1,np.abs(dft(y)), 'b',markerfmt=" ", basefmt="-b",use_line_collection="True")


plt.show()
```
